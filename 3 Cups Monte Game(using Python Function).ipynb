### Rules of the game :

The game has one very simple rule. And that is to correctly guess the red ball in only one of the three cups The prerequisite is just not knowing clearly what is in these cups before guessing. Thatâ€™s all.

```python
# we need help to shuffle our cups. let's call a library to shuffle our cups by importing them

from random import shuffle
```

```python
list_ex = [1,2,3]

shuffle(list_ex)

list_ex

#shuffled list example
```

```python
#now mimic our cup by making list and balls within

mycups = [" ","Balls"," "]

#create a function to shuffling our cups

def shuffle_cups(my_cups):
    shuffle(my_cups)
    return (my_cups)
```

```python
shuffle_cups(mycups)

#our cups is shuffled succesfully!
```

```python
# the next step is creating "player guess" function

def player_guess():
    guess = " " #start with empty guess
    
    #when guess is not within the cups range option, keep asking
    while guess not in ["0","1","2"]:
        guess = input("pick a cups number 0 , 1 or 2 ? ")
        
    return int(guess) # needs an "int" since the returned data will be "str" in default
                      # the guess is needed to indexing the list, so we need "int" value

player_guess()
```

```python
#last steps, checking guess function 

def check_guess(mycups,guess):
    
    
    if mycups[guess] == "Balls":
        print ("Correct Guess! YOU WON THE GAME !! ")
        print(mycups)
    else: 
        print ("Better luck next time buddy")
        print(mycups)
        
```

```python
#combining the function to create the game (this is the most exciting part !)

#initial cups 
mycups = [" ","Balls"," "]

#shuffle cups
shuffling = shuffle_cups(mycups)

#player guess
guess = player_guess()

#checking_guess
check_guess(shuffling,guess)
```
